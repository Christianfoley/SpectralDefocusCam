```python
# define fwd model
mask = load_mask()
num_images = len(fruit_train_data)
num_inputs_simult = 6
blur_type = 'symmetric'
optimize_blur = False

forward_model = fm.Forward_Model(mask, num_ims = num_inputs_simult, blur_type = blur_type, optimize_blur = optimize_blur)
recon_model = Unet3d.Unet(n_channel_in=num_inputs_simult, n_channel_out=1) #Unet or Unet3d.Unet
recon_model.to(device)
model = sm.MyEnsemble(forward_model.to(device), recon_model.to(device))
```

```python
def show_false_color(img, k = 40):
    plt.imshow(np.rot90(helper.stack_rgb_opt_30(img, scaling = [0.49*k,0.837*k,2.02*k], channels = 30),k=3))
    plt.axis('off')
    plt.show()

loss_fn_alex = lpips.LPIPS(net='alex') # best forward scores
loss_fn_vgg = lpips.LPIPS(net='vgg') # closer to "traditional" perceptual loss, when used for optimization
loss_function = MSELoss() #CosineSimilarity()
loss_function_2 = CosineSimilarity(dim = 1)

def get_lpips_loss(img1, img2):
    lpips_loss = []
    for i in range(6,7):#img1.shape[1]//3):
        a = img1[:,3*i:3*i + 3,:,:].detach().cpu()
        b = img2[:,3*i:3*i + 3,:,:].detach().cpu()
        lpips_loss.append(loss_fn_vgg(a, b).item())
    return np.mean(np.asarray(lpips_loss))
```
